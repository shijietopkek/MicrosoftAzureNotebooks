```python
'''
The Fibonacci sequence was originally used as a basic model for rabbit population growth.
The Fibonacci sequence goes like this:

0,1,1,2,3,5,8,13,...

The nth term is the sum of the previous two terms.
In the above example, the next term would be 21 (8+13).

Create a Fibonacci sequence generator up to the 50th term. 
The program will then prompt the user for n (up to 50).
E.g
Enter n: 7
Fibonacci number is 8
'''

arr = [None for i in range(50)]
arr[0] = 0
arr[1] = 1
def fibonacci(num):
    for i in range(2,num):
        arr[i] = arr[i-1] + arr[i-2]

#main
fibonacci(50)
user = int(input("Enter a number between 1 and 50: "))
if user <= 0:
    print("Too low")
elif user > 50:
    print("Too high")
else:
    print("Fibonacci number is {}".format(arr[user-1]))
    
```

```python
'''
Create a simple treasure hunt game.
Create a two-dimensional 10 by 10 array of integers.
In a random position in the array store the number 1.
Repeat
  Get the user to enter coordinates where he/she think the treasure is.
  If there is a 1 at this position display 'Bingo!'.
Until they find the treasure
'''
import random
arr = [[None for i in range(10)] for j in range(10)]

arr[random.randint(0,9)][random.randint(0,9)] = 1
while True:
    user = input("Enter coordinates: ")
    row,column = user.split(" ")
    row = int(row)
    column = int(column)
    if arr[row][column]==1:
        print("Bingo!")
        break;
    
    
        
    
```

```python
'''
For the above treasure hunt game, add a feature to say
'hot', 'cold' or 'warm'
depending on how close the guess was to the location. 
'''
import random
arr = [[None for i in range(10)] for j in range(10)]

x = random.randint(0,9)
y = random.randint(0,9)
arr[x][y] = 1
while True:
    user = input("Enter coordinates: ")
    row,column = user.split(" ")
    row = int(row)
    column = int(column)
    if arr[row][column]==1:
        print("Bingo!")
        break;
    else:
        dist_from = (abs(x-row)**2 + abs(y-column)**2)**0.5
        if dist_from < 3:
            print("Hot")
        elif dist_from < 5:
            print("Warm")
        else:
            print("Cold")
    
    
```

```python
'''
The computer generates a 4-digit code
The user types in a 4-digit code. His/her guess.
The computer tells them how many digits they guessed correctly
'''
import random

def match(str1,str2):
    count = 0
    if len(str1) != len(str2):
        return 0,False
    else:
        arr = []
        for i in range(len(str1)):
            if str1[i] == str2[i]:
                count+=1
        if count==4:
            return count,True
        return count,False
        



computer = random.randint(1000,9999)
computer = str(computer)
while True:
    user = int(input("Enter a 4-digit code: "))
    count,condition = match(computer,str(user))
    print("Number of correct digits: {}".format(count))
    if condition:
        print("Bingo!")
        break;
    
    

```

```python
'''
For the above 4-digit code game, the computer tells them how many 
digits are guessed correctly in the correct place and how many 
digits have been guessed correctly but in the wrong place.
The user gets 12 guesses to either win â guess the right code, 
or lose â run out of guesses. 
'''
import random

def match(str1,str2):
    count = 0
    count1 = 0
    if len(str1) != len(str2):
        return 0,0,False
    else:
        arr = []
        for i in range(len(str1)):
            if str1[i] == str2[i]:
                count+=1
            else:
                arr.append(str2[i])
        for i in set(arr):
            if i in str1:
                count1+=1
        if count==4:
            return count,count1,True
        return count,count1,False
        



computer = random.randint(1000,9999)
computer = str(computer)
num = 0
win = False
while num<12:
    user = int(input("Enter a 4-digit code: "))
    correct_pos, num_exist, condition = match(computer,str(user))
    print("Correct number, correct place: {}\nCorrect number, wrong place: {}".format(correct_pos, num_exist))
    if condition:
        print("You win!")
        win = True
        break;
    num+=1

if not win:
    print("You lose")
```

```python
'''
Player 1 types in a word
Player 2 has to guess the word in 5 lives
The display would look like this:

Word to guess: ******
You have 5 lives left - Letter? E
Word to guess: **EE*E
You have 5 lives left - Letter? Z
Word to guess: **EE*E
You have 4 lives left - Letter?
'''


def new_format(word_to_guess, actual_word, letter):
    condition = 2
    word_to_guess = list(word_to_guess)
    for i in range(len(actual_word)):
        if actual_word[i] == letter and word_to_guess[i] != letter:
            condition = 3
            word_to_guess[i] = letter
    word_to_guess ="".join(word_to_guess)
    if word_to_guess.count("*")==0:
        condition = 1
    return word_to_guess, condition
no_of_lives = 5
player_1 = input()
word_to_guess = "*" * len(player_1)
print("Word to guess: {}".format(word_to_guess))
while no_of_lives > 1:
    player_2 = input("You have {} lives left - Letter? ".format(no_of_lives))
    word_to_guess, condition = new_format(word_to_guess,player_1,player_2)
    if condition==1:
        print("Word to guess: {}".format(word_to_guess))
        print("You win!")
        break;
    elif condition==3:
        print("Word to guess: {}".format(word_to_guess))
    else:
        no_of_lives -=1
        print("Word to guess: {}".format(word_to_guess))
else:
    print("You lose.")








```
